```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
prices = pd.read_csv("stock.csv", parse_dates=["Date"])
```

```python
prices
```

```python
stock = pd.DataFrame()
```

```python
stock["Date"] = prices["Date"]
stock["Apple"] = prices["Adj Close Price"]
```

```python
stock.head()
```

```python
stock["Apple"].head(30).mean(), stock["Apple"].iloc[1:31].mean()
```

```python
stock["SMA30"] = stock["Apple"].rolling(30).mean()
```

```python
pd.options.display.max_rows = None
```

```python
stock
```

```python
stock["SMA100"] = stock["Apple"].rolling(100).mean()
```

```python
stock
```

```python
for i, row in stock.iterrows():
    print(row["SMA30"])
```

```python
def buy_sell(df):
    buy_price_price = []
    sell_price_price = []
    holding = [] # 1 for hold, 0 for not hold
    actions = [] # 1 for buy, -1 for sell, 0 for wait
    sig = 0
    for i,row in df.iterrows():
        if row["SMA30"] > row["SMA100"]: # buy
            if sig != 1:
                buy_price_price.append(row["Apple"])
                sell_price_price.append(np.nan)
                sig = 1
                actions.append(1)
                holding.append(1)
            else:
                buy_price_price.append(np.nan)
                sell_price_price.append(np.nan)
                actions.append(0)
                holding.append(1)
        elif row["SMA30"] < row["SMA100"]: # sell
            if sig != -1:
                buy_price_price.append(np.nan)
                sell_price_price.append(row["Apple"])
                sig = -1
                actions.append(-1)
                holding.append(0)
            else:
                buy_price_price.append(np.nan)
                sell_price_price.append(np.nan)
                actions.append(0)
                holding.append(0)
        else:
            holding.append(0)
            buy_price_price.append(np.nan)
            sell_price_price.append(np.nan)
            actions.append(0)
            
    return buy_price_price, sell_price_price, holding, actions
```

```python
buysell = buy_sell(stock)
```

```python
stock["Buy Price"], stock["Sell Price"], stock["Hold"], stock["Actions"] = buysell
```

```python
plt.figure(figsize=(16,5))
plt.style.use("dark_background")

plt.plot(stock["Date"], stock["Apple"], label = "Apple", alpha = 0.35)
plt.plot(stock["Date"], stock["SMA30"], label = "SMA30", alpha = 0.35)
plt.plot(stock["Date"], stock["SMA100"], label = "SMA100", alpha = 0.35)

plt.scatter(stock["Date"], stock["Buy Price"], color = "green", s = 100, marker = "^")
plt.scatter(stock["Date"], stock["Sell Price"], color = "red", s = 100, marker = "v")

plt.title("Apple Historical Adj Close Price",weight = "bold")
plt.xlabel("From May, 2014 to 22 May, 2020",weight = "bold")
plt.ylabel("Adj Close Price",weight = "bold")

plt.show()
```
